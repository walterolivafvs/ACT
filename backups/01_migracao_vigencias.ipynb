```python
import pandas as pd

# ==========
# 1) Ler planilha NOVA (a sua tabela estruturada)
# ==========
nova_path = "planilha_nova.xlsx"
df_nova = pd.read_excel(nova_path, sheet_name="tbl_instrumentos")

# garantir coluna id_num na nova
df_nova["id_num"] = (
    df_nova["identificacao"].astype(str)
    .str.extract(r"(\d+/\d{4})", expand=False)
)

# ==========
# 2) Ler planilha ANTIGA (com header correto)
# ==========
antiga_path = "controle_acts_e_convenios.xlsx"
df_antiga = pd.read_excel(antiga_path, sheet_name="TCT", header=2)

# limpar nomes das colunas (remove espaços invisíveis)
df_antiga.columns = [str(c).strip() for c in df_antiga.columns]

# criar id_num na antiga
df_antiga["id_num"] = (
    df_antiga["IDENTIFICAÇÃO"].astype(str)
    .str.extract(r"(\d+/\d{4})", expand=False)
)

# converter datas (formato BR)
df_antiga["VIGÊNCIA - INCIO"] = pd.to_datetime(df_antiga["VIGÊNCIA - INCIO"], errors="coerce", dayfirst=True)
df_antiga["VIGÊNCIA - TÉRMINO"] = pd.to_datetime(df_antiga["VIGÊNCIA - TÉRMINO"], errors="coerce", dayfirst=True)

# ==========
# 3) Deduplicar antiga por id_num (se existir duplicado, escolhe a linha com maior término)
# ==========
df_antiga_dedup = (
    df_antiga.sort_values("VIGÊNCIA - TÉRMINO")
             .drop_duplicates(subset=["id_num"], keep="last")
)

# ==========
# 4) Merge completo
# ==========
df_merge = df_nova.merge(
    df_antiga_dedup[["id_num", "VIGÊNCIA - INCIO", "VIGÊNCIA - TÉRMINO"]],
    on="id_num",
    how="left"
)

df_merge["vigencia_inicio"] = df_merge["VIGÊNCIA - INCIO"]
df_merge["vigencia_termino"] = df_merge["VIGÊNCIA - TÉRMINO"]

# ==========
# 5) Checagens rápidas
# ==========
print("Preenchimento (notna):")
print(df_merge[["vigencia_inicio", "vigencia_termino"]].notna().sum())

print("\nSem id_num (na planilha nova):")
print(df_merge[df_merge["id_num"].isna()][["id_instrumento", "identificacao"]])

print("\nSem vigencia_termino (mesmo com id_num):")
print(df_merge[(df_merge["id_num"].notna()) & (df_merge["vigencia_termino"].isna())][["id_instrumento", "identificacao", "id_num"]])
```

```python
def texto_para_bullets(txt):
    if pd.isna(txt) or str(txt).strip() == "":
        return ""
    linhas = [l.strip() for l in str(txt).split("\n") if l.strip()]
    return "<ul>" + "".join(f"<li>{l}</li>" for l in linhas) + "</ul>"
```

```python
cols_bullets = [
    "objetivo_resumido",
    "produtos_gerados"
]
```

```python
for c in cols_bullets:
    df[c] = df[c].apply(texto_para_bullets)
```
